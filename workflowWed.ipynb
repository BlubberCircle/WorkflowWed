```python
%matplotlib inline
import pandas as pd

import seaborn as sns

import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r"C:\TEMP\Attributes.csv")
```

```python
data.head()
```

```python
data.describe()
```

```python
data = data[data.columns.drop(list(data.filter(regex='UTM')))]

data = data.dropna()
```

```python
g = sns.PairGrid(data, diag_sharey=False)
g.map_lower(sns.kdeplot)
g.map_upper(sns.regplot,fit_reg=True)
g.map_diag(sns.kdeplot, lw=3)
```

```python
corr = data.corr()
```

```python
f, ax = plt.subplots(figsize=(16,16))

hm = sns.heatmap(round(corr,2),  annot=True, cmap="coolwarm",fmt='.2f',linewidths=.05)

f.subplots_adjust(top=0.93)
ax.set_ylim(6, -0.5)

plt.setp(ax.get_yticklabels(), rotation=15, ha="right", rotation_mode="anchor");
```
